## LIBRARIES

```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inside the notebook
%matplotlib inline

# Set plot style
sns.set(style="whitegrid")
```

## Load Dataset

```python
df=pd.read_csv("day.csv")
df
```

## Explore Basic Info

```python
# Shape of the dataset
print("Shape:", df.shape)

# Check data types and null values
df.info()

# Summary statistics
df.describe()
```

## Convert Dates and Add Useful Columns

```python
# Convert date column to datetime
df['dteday'] = pd.to_datetime(df['dteday'])
df.info()
```

```python
# Extract day name and month for grouping later
df['day_of_week'] = df['dteday'].dt.day_name()
df['month'] = df['dteday'].dt.month_name()
```

## Rentals by Day of Week

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='day_of_week', y='cnt', estimator= sum, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title("Total Rentals by Day of the Week")
plt.xlabel("Day of Week")
plt.ylabel("Total Rentals")
plt.tight_layout()
plt.show()
```

```python
df.columns
```

## Temperature vs Rentals

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='temp', y='cnt', data=df)
plt.title("Temperature vs Total Rentals")
plt.xlabel("Normalized Temperature")
plt.ylabel("Total Rentals")
plt.tight_layout()
plt.show()
```

## Weather Situation vs Rentals

```python
# Weather Situations: 1 = Clear, 2 = Mist/Cloudy, 3 = Light Snow/Rain, 4 = Heavy Rain/Snow
weather_map = {
    1: 'Clear',
    2: 'Cloudy',
    3: 'Light Snow/Rain',
    4: 'Heavy Rain/Snow'
}

df['weathersit'] = df['weathersit'].map(weather_map)

plt.figure(figsize=(8, 5))
sns.boxplot(x='weathersit', y='cnt', data=df)
plt.title("Rentals vs Weather Situation")
plt.xlabel("Weather Condition")
plt.ylabel("Total Rentals")
plt.tight_layout()
plt.show()
```

## Casual vs Registered Users


```python
# Total counts of casual and registered users
casual_total = df['casual'].sum()
registered_total = df['registered'].sum()

plt.figure(figsize=(6, 4))
sns.barplot(x=['Casual', 'Registered'], y=[casual_total, registered_total])
plt.title("Total Users: Casual vs Registered")
plt.ylabel("User Count")
plt.tight_layout()
plt.show()
```

## Monthly Trend

```python
monthly = df.groupby('month')['cnt'].sum().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])

monthly.plot(kind='bar', figsize=(12, 5), color='teal')
plt.title("Monthly Total Rentals")
plt.ylabel("Total Rentals")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
